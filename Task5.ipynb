1. What is the age distribution of people with and without heart disease?

Solution 1: Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram
sns.histplot(data=df, x='age', hue='target', kde=True, multiple="stack")
plt.title('Age Distribution by Heart Disease')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

Solution 2: Box Plot

```python
# Box Plot
sns.boxplot(data=df, x='target', y='age')
plt.title('Age Distribution by Heart Disease')
plt.xlabel('Heart Disease')
plt.ylabel('Age')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

2. What is the distribution of resting blood pressure among people with and without heart disease?

Solution 1: Histogram

```python
# Histogram
sns.histplot(data=df, x='trestbps', hue='target', kde=True, multiple="stack")
plt.title('Resting Blood Pressure Distribution by Heart Disease')
plt.xlabel('Resting Blood Pressure')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

Solution 2: Violin Plot

```python
# Violin Plot
sns.violinplot(data=df, x='target', y='trestbps')
plt.title('Resting Blood Pressure Distribution by Heart Disease')
plt.xlabel('Heart Disease')
plt.ylabel('Resting Blood Pressure')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

3. What is the relationship between age and maximum heart rate achieved among people with and without heart disease?

Solution 1: Scatter Plot

```python
# Scatter Plot
sns.scatterplot(data=df, x='age', y='thalach', hue='target')
plt.title('Age vs. Max Heart Rate by Heart Disease')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

Solution 2: Line Plot

```python
# Line Plot
sns.lineplot(data=df, x='age', y='thalach', hue='target', ci=None)
plt.title('Age vs. Max Heart Rate by Heart Disease')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

4. What is the distribution of cholesterol levels among people with different types of chest pain?

Solution 1: Box Plot

```python
# Box Plot
sns.boxplot(data=df, x='cp', y='chol', hue='target')
plt.title('Cholesterol Levels by Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Cholesterol Level')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

Solution 2: Violin Plot

```python
# Violin Plot
sns.violinplot(data=df, x='cp', y='chol', hue='target', split=True)
plt.title('Cholesterol Levels by Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Cholesterol Level')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

5. How does the resting electrocardiographic results distribution look among people with and without heart disease?

Solution 1: Count Plot

```python
# Count Plot
sns.countplot(data=df, x='restecg', hue='target')
plt.title('Resting Electrocardiographic Results by Heart Disease')
plt.xlabel('Resting ECG Results')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

Solution 2: Bar Plot

```python
# Bar Plot
df['count'] = 1
restecg_heart_disease = df.groupby(['restecg', 'target']).count()['count'].unstack().fillna(0)
restecg_heart_disease.plot(kind='bar', stacked=True)
plt.title('Resting Electrocardiographic Results by Heart Disease')
plt.xlabel('Resting ECG Results')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```
